```python
# Automatically count classes
import os

label_dir = "/content/retail/train/labels"
class_ids = set()

for file in os.listdir(label_dir):
    with open(os.path.join(label_dir, file)) as f:
        for line in f:
            if line.strip():
                class_ids.add(int(line.split()[0]))

print("Number of classes:", len(class_ids))
```

```python
import yaml

with open("/content/retail/data.yaml", "r") as f:
    data = yaml.safe_load(f)

num_classes = data["nc"]
class_names = data["names"]

print("Number of classes:", num_classes)
print("Class names:", class_names)
```

```python
import os

image_dir = "/content/retail/train/images"
label_dir = "/content/retail/train/labels"

images = set(os.path.splitext(f)[0] for f in os.listdir(image_dir) if f.endswith('.jpg'))
labels = set(os.path.splitext(f)[0] for f in os.listdir(label_dir) if f.endswith('.txt'))

matched = images & labels
print("Images:", len(images))
print("Labels:", len(labels))
print("Matched pairs:", len(matched))
```

```python
data_yaml = f"""\
path: /content/retail
train: train/images
val: valid/images
test: test/images
nc: {num_classes}
names: {class_names}
"""

with open("retail.yaml", "w") as f:
    f.write(data_yaml)
```

```python
model = YOLO("yolov8l.pt")
```

```python
model.info(detailed=True)
```

```python
model.train(
    data="retail.yaml",
    epochs=20,
    imgsz=640,
    batch=16,
    name="head_only",
    project="/content/drive/MyDrive/Colab Notebooks/Smart-retail-checkout-system/yolo_runs",  # <– save to Drive
    patience=10,                # ← early stopping
    save=True,                  # ← saves best.pt automatically
    save_period=1,              # ← save checkpoint every epoch
    val=True,                    # ← validate after every epoch
    #freeze=177                   # freeze backbone; head remains trainable
)
```

```python
# Resume training
model = YOLO("/content/drive/MyDrive/Colab Notebooks/Smart-retail-checkout-system/yolo_runs/head_only/weights/last.pt")
model.train(resume=True)
```

```python
# Evaluate on validation data
metrics = model.val()
print(metrics)
```

```python
# Evaluate on test data
metrics = model.val(split='test')
```

```python
results = model.predict(
    source="/content/retail/test/images",
    conf=0.5,
    save=True
)
```

```python
model.export(format="onnx")
model.export(format="torchscript")
model.export(format="tflite")
```

```python
# Access Predicted Data Programmatically
for r in results:
    boxes = r.boxes
    for box in boxes:
        print("Class:", model.names[int(box.cls)])
        print("Confidence:", box.conf.item())
        # Move the tensor to the CPU before converting to NumPy
        print("Box coordinates:", box.xyxy.cpu().numpy())
```
